```python
def missing_vals(df):
    missing_data_df = []
    missing_data_df = pd.DataFrame(df.isnull().sum())
    missing_data_df.columns = ['missing_values']
    missing_data_df['missing%'] = missing_data_df['missing_values']/df.shape[0]
    return missing_data_df
```

```python
import geopandas as gpd
from shapely.geometry import Point, Polygon

%matplotlib inline
```

```python
geometry = [Point(xy) for xy in zip(events_df.longitude, events_df.latitude)]
```

```python
geometry[:5]
```

```python
crs = {'init': 'epsg:4326'}
event_locs = gpd.GeoDataFrame(events_df, crs=crs, geometry=geometry)

event_locs.head()
```

```python
fig, ax = plt.subplots(figsize = (8,8))
event_locs.geometry.plot(marker='d', color='blue', markersize=1, ax=ax, label = 'Telecom Use Loc')

plt.title('INSAID Telecom Usage areas')
plt.legend()
```
